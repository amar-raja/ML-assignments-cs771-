```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#GIVEN DATA
n=50     #NUMBER OF OBSERVATIONS
mean=10
sd=5
data=np.random.normal(mean,sd,n)        #GENERATING 50 DATA POINTS IN THE DISTRIBUTION
data
```

$$
\begin{aligned}
LIKELIHOOD\ DISTRIBUTION=>
N\left(\bar{y}, \frac{\sigma^{2}}{n}\right)
\end{aligned}
$$

```python
#mean and sd for likelihood
mean_lh = np.mean(data)
sd_lh = np.sqrt(sd**2/n)
```

```python
#Given mean and sd for prior
mean_prior= 25
sd_prior= 5
```

$$
\begin{aligned}
VARIANCE => \sigma_{n}^{2} = \frac{\sigma_{l}^{2}{s^{2}}}{\sigma_{l}^{2}+{s^{2}}},
\ MEAN=>
\mu_n=\frac{m\sigma_l^2+\bar{y}s^2}{\sigma_l^2+s^2}
\end{aligned}
$$

```python
#mean and sd for posterior

mean_post = (mean_prior*(sd_lh**2)+mean_lh*(sd_prior**2))/((sd_lh**2)+(sd_prior**2))
sd_post = np.sqrt(((sd_lh**2)*(sd_prior**2))/((sd_lh**2)+(sd_prior**2)))
```

```python
#FUNCTION TO PLOT PDF
def pdf(data):
    sd=np.sqrt(np.var(data))
    mean=np.mean(data)
    val = 1/(sd*np.sqrt(2*np.pi))*np.exp(-(data-mean)**2/(2*(sd**2)))
    return val
```

```python
# GENERATING SAMPLE DATA

data_lh = np.random.normal(mean_lh,sd_lh,500)
data_prior = np.random.normal(mean_prior,sd_prior,500)
data_post = np.random.normal(mean_post,sd_post,500)
```

```python
#SORTING X_VALUES
data_lh.sort()
data_prior.sort()
data_post.sort()
```

```python
#PLOTTING DISTRIBUTIONS OF LIKELIHOOD,PRIOR,POSTERIOR
val_lh = pdf(data_lh)
val_prior=pdf(data_prior)
val_post=pdf(data_post)
```

```python
plt.title('pdfs of likelihood,prior and posterior')
plt.plot(data_lh,val_lh,label="likelihood",color='r')
plt.plot(data_post,val_post,label="posterior",color='g')
plt.plot(data_prior,val_prior,label="prior",color='b')
plt.legend()
```

### DERIVATION OF LIKELIHOOD

Likelihood of $n$ observations 
Given iid $N\left(\mu, \sigma^{2}\right) ; \sigma^{2}$ is a known constant
$$
\begin{aligned}
f\left(y_{1}, \ldots, y_{n} \mid \mu, \sigma^{2}\right) &=\prod_{i=1}^{n} \frac{1}{\sqrt{2 \pi\sigma^{2}}} \exp \left\{\frac{-1}{2 \sigma^{2}}\left(y_{i}-\mu\right)^{2}\right\} \\
&=\left(\frac{1}{\sqrt{2 \pi\sigma^{2}}}\right)^{n} \exp \left\{-\frac{1}{2 \sigma^{2}} \sum_{i=1}^{n}\left(y_{i}-\mu\right)^{2}\right] \\
& \propto \exp \left\{-\frac{1}{2 \sigma^{2}} \sum_{i=1}^{n}\left(y_{i}^{2}+\mu^{2}-2 \mu y_{i}\right)\right\} \\
&=\exp \left\{-\frac{1}{2 \sigma^{2}}\left(\sum_{i=1}^{2} y_{i}^{2}+n {\mu}^{2}-2 \mu n \bar{y}\right)\right\} \\
&=\exp \left\{-\frac{n}{2 \sigma^{2}}\left(\mu^{2}-2 \mu \bar{y}+\frac{1}{n} \sum_{i=1}^{n} y_{i}^{2}\right)\right\} \\
\end{aligned}
$$

Comparing it with $\exp \left\{-\frac{1}{2 \sigma_{l}^{2}}\left(\mu-\mu_{l}\right)^{2}\right\}$ ;where $\mu_{l}$ and $\sigma_{l}$ are mean and standard deviation of likelihood
$$
\begin{aligned}
&=\exp \left\{-\frac{1}{2 \sigma_{l}^{2}}\left(\mu^{2}-2 \mu \mu_{l}+\mu_{l}^{2}\right)\right\}\\
&=> \mu_{l} = \bar{y} ; -\frac{n}{2\sigma^{2}}=-\frac{1}{2\sigma_{l}^{2}} \\
&=> \sigma_{l}^{2} = \frac{\sigma^{2}}{n}\\
&=>N\left(\bar{y}, \frac{\sigma^{2}}{n}\right)
\end{aligned}
$$

### DERIVATION OF POSTERIOR

Likelihood, n iid samples from a normal $N\left(\mu, \sigma^{2}\right)$ is $N\left(\bar{y}, \frac{\sigma^{2}}{n}\right)$ 
Let $\mu_{l}=\bar{y},\sigma_{l}^2=\frac{\sigma^{2}}{n}$ and Prior is $N\left(m, s^{2}\right)$ 
then Posterior $\alpha$ Prior $\times$ Likelihood will be $N\left(\mu_{n}, \sigma_{n}^{2}\right)$ 
As we know $s^{2}$ and $\sigma^{2}$, we can ignore the constent term
$$
\begin{aligned}
&\frac{1}{\left(\sqrt{2 \pi \frac{\sigma^{2}}{n}}\right)^n} \times \frac{1}{\sqrt{2 \pi s^{2}}} \\
&g\left(\mu \mid y_{1}, \ldots, y_{n}\right) \propto g(\mu) \times f\left(y_{1}, \ldots, y_{n} \mid \mu\right)\\
&\alpha \exp \left\{-\frac{1}{2 s^{2}}(\mu-m)^{2}\right\} \times \exp \left\{-\frac{n}{2 \sigma^{2}}(\mu-\bar{y})^{2}\right\}\\
&=\exp \left\{-\frac{1}{2 s^{2}}\left(\mu^{2}+m^{2}-2 \mu m\right)-\frac{n}{2 \sigma^{2}}\left(\mu^{2}+\bar{y}^{2}-2 \mu \bar{y}\right)\right\}\\
&=\exp \left\{-\frac{1}{2}\left[\mu^{2}\left(\frac{1}{s^{2}}+\frac{n}{\sigma^{2}}\right)-2 \mu\left(\frac{m}{s^{2}}+\frac{n \bar{y}}{\sigma^{2}}\right)+\frac{m^{2}}{s^{2}}+\frac{\bar{y}^{2}}{\sigma^{2}}\right]\right\}\\
\end{aligned}
$$
Comparing the above with the required: $\exp \left\{-\frac{1}{2 \sigma_{n}^{2}}\left[\mu^{2}+\mu_{n}^{2}-2 \mu \mu_{n}\right]\right\}$
$$
\begin{aligned}
&-\frac{1}{2 \sigma_{n}^{2}} \mu^{2}=\frac{-\mu^{2}}{2}\left(\frac{1}{s^{2}}+\frac{n}{\sigma^{2}}\right) \Rightarrow \frac{1}{\sigma_{n}^{2}}=\left(\frac{1}{s^{2}}+\frac{n}{\sigma^{2}}\right) \\
&\frac{1}{\sigma_{n}^{2}}=\frac{\sigma^{2}+s^{2} n}{\sigma^{2} s^{2}} \implies \sigma_{n}^{2}=\frac{\sigma^{2} s^{2}}{\sigma^{2}+s^{2} n} \\
&\implies \sigma_{n}^{2} = \frac{\sigma_{l}^{2}{s^{2}}}{\sigma_{l}^{2}+{s^{2}}} \\
\end{aligned}
$$

Similarly,
$$
\begin{aligned}
&\frac{2\mu\mu_{n}}{2\sigma_{n}^{2}}=\frac{\mu\mu_{n}}{\sigma_{n}^{2}} =\mu\left(\frac{m}{s^{2}}+\frac{n\bar{y}}{\sigma^{2}}\right) \\
&\frac{\mu_{n}}{\sigma_{n}^2}=\frac{m\sigma^2+n\bar{y}s^2}{\sigma^{2}s^2} \\
&\mu_n = \sigma_n^2\left(\frac{m\sigma^2+n\bar{y}s^2}{\sigma^2s^2}\right)=\frac{\sigma^{2} s^{2}}{\sigma^{2}+s^{2} n}\left(\frac{m\sigma^2+n\bar{y}s^2}{\sigma^2s^2}\right) \\
&=\frac{m\sigma^2+n\bar{y}s^2}{\sigma^2+s^2 n}=\frac{m\frac{\sigma^2}{n}+\bar{y}s^2}{\frac{\sigma^2}{n}+s^2} \\
&\implies \mu_n=\frac{m\sigma_l^2+\bar{y}s^2}{\sigma_l^2+s^2} \\
\end{aligned}
$$


# 2B)METROPOLIS ALGORITHM

#### ANALYSIS OF FROM MNMC AND ANALYTIC POSTERIOR DISTRIBUTION

```python
#FUNCTION TO PLOT PDF FOR EACH VALUE MEAN CENTERED
def pdf_1(x,data):
    sd=np.sqrt(np.var(data))
    mean=np.mean(data)
    val = 1/(sd*np.sqrt(2*np.pi))*np.exp(-(x-mean)**2/(2*(sd**2)))
    return val
```

```python
#RETURNING PROBABILITY OF x IN PRIOR DISTRIBUTION
def prior(x):
    return pdf_1(x,data_prior)
```

```python
#RETURNING PROBABILITY OF x IN LIKELIHOOD DISTRIBUTION
def likelihood(x):
    #data2 is sampled from likelihood distribution
    
    return pdf_1(x,data_lh)
```

```python
#TRANSITION MODEL TO GENERATE THE NEXTSTATE FROM THE PREVIOUS STATE
#TRANSITION MODEL DISTRIBUTION IS NORMAL DISTRIBUTION

def transition_model(u,sd):
    #returning NEXTSTATE
    
    return np.random.normal(u,sd,1)
```

```python
#ACCEPTANCE FUNCTION

def acceptance(x_posterior,x_new_posterior):
    r = x_new_posterior/x_posterior
    
    if r>=1:
        return True
    else:
        #generate random num between 0 and 1
        
        accept = np.random.uniform(0,1)
        
        #return true if that num is less than r else false
        
        return (accept < r)
```

```python
def metropolis_hastings(param_init,iterations,sd_for_transition_model):
    #initializing values
    
    x=param_init
    accepted = []
    rejected = []
    accepted_iter = []
    
    count=0
    ls=[]
    for i in range(iterations):
        
        #likelihood prob
        x_lh = likelihood(x)
        
        #prior probability
        x_prior = prior(x)
        
        x_posterior = x_lh *x_prior
        
        #new x from transition model
        x_new = transition_model(x,sd_for_transition_model)
        
        #likelihood prob for x_new
        
        x_new_lh = likelihood(x_new)
        
        x_new_prior = prior(x_new)
        
        x_new_post = x_new_lh * x_new_prior
        
        #ACCPETANCE CHECKING
        
        if(acceptance(x_posterior,x_new_post)):
            if(count >200):
                print('TERMINATED')
                return np.array(accepted) , np.array(rejected) , np.array(accepted_iter)
            x=x_new
            accepted.append(x_new)
            accepted_iter.append(i)
            count+=1
            ls.append(count)
            
        else:
            #ADDING X_NEW TO REJECT LIST
            count = 0
            rejected.append(x_new)
    return np.array(accepted) , np.array(rejected) , np.array(accepted_iter)
```

```python
accepted , rejected , accepted_iter = metropolis_hastings(10,50000,1)
```

```python
# speed of convergence of sampling dependency on the proposal width (STANDARD DEVIATION)

std_ls = [0.8,0.9,1.0,1.1,1.2,1.3,1.4,1.5,5]
for i in std_ls:
    accepted , rejected , accepted_iter = metropolis_hastings(10,5000,i)

    plot3 = plt.figure(3)
    
    bin_centers1 = 0.5*(bins1[1:] + bins1[:-1])
    den1 = pdf_1(bin_centers1,data_post)
    plt.plot(bin_centers1,den1,label='analytic_aposterior',color='r')
    
    bin_centers2 = 0.5*(bins2[1:] + bins2[:-1])
    den2 = pdf_1(bin_centers2,accepted)
    plt.plot(bin_centers2,den2,label='mnmc_posterior',color='b')
    plt.title('FOR STANDARD DEVIATION:{}'.format(i))
    plt.legend()
    plt.show()
```

# FROM ABOVE WE HAVE THE OPTIMAL STANDARD DEVIATION =1.0 FOR 5000 ITERATI0NS

```python
#NOW WE DO IT FOR DIFFERENT NUMBER OF ITERATIONS
inter_ls = [2500,5000,7500,10000,12500,15000]
for i in inter_ls:
    accepted , rejected , accepted_iter = metropolis_hastings(10,i,1.0)

    plot3 = plt.figure(3)
    
    bin_centers1 = 0.5*(bins1[1:] + bins1[:-1])
    den1 = pdf_1(bin_centers1,data_post)
    plt.plot(bin_centers1,den1,label='analytic_aposterior',color='r')
    
    bin_centers2 = 0.5*(bins2[1:] + bins2[:-1])
    den2 = pdf_1(bin_centers2,accepted)
    plt.plot(bin_centers2,den2,label='mnmc_posterior',color='b')
    plt.title('FOR SD:1.0 and ITERATIONS :{}'.format(i))
    plt.legend()
    plt.show()
```

## FROM ABOVE WE CAN SEE THAT AS NUMBER OF ITERATIONS ARE INCREASED IT CONVERGES OPTIMALLY
